```python
from ipywidgets import interactive
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


%matplotlib inline
```

```python
def conv_int(x):
    try:
        return int(x)
    except ValueError as ve:
        print('%s for %s' % (ve, x))
        pass

df_rev = pd.read_csv('sdf17_1a.txt', sep='\t')

# df_rev[df_rev['LEAID'] == '09D0001']

df_rev['LEAID'] = df_rev['LEAID'].apply(conv_int)

df_rev['LEAID'] = df_rev['LEAID'].fillna(0).astype(int)

df_rev = df_rev.groupby(['STABBR', 'NAME', 'LEAID'])[['TOTALREV']].sum()

df_rev.index = df_rev.index.set_names('ST', level=0)

df_rev.loc[('MI'), :]
```

```python
x = np.array(df_rev.index.get_level_values(0))
lst_states = np.unique(x)

for value in lst_states:
    print('%s: %s' % (value, str(len(df_rev.loc[value]))))
```

```python
df_mem = pd.read_csv('membership.csv',  sep=',')
df_mem = df_mem[(~df_mem['DMS_FLAG'].isin(['Not reported', 'Missing'])) & (df_mem['RACE_ETHNICITY'] != 'No Category Codes')]
df_mem = df_mem.groupby(['ST', 'LEA_NAME', 'LEAID', 'RACE_ETHNICITY'])["STUDENT_COUNT"].sum()
df_mem = df_mem.unstack(level=-1)
df_mem['Total'] = df_mem.sum(axis=1)
# df_mem.index = df_mem.index.set_names('NAME', level=1)
```

```python
df_mem.loc[('MI'), :]
col_mem = list(df_mem.columns)
col_mem.remove('White')
col_mem.remove('Total')
print(list(col_mem))
```

```python
x = np.array(df_mem.index.get_level_values(0))
lst_states = np.unique(x)

for value in lst_states:
    print('%s: %s' % (value, str(len(df_mem.loc[value]))))
```

```python
df_com = df_mem.merge(df_rev, left_index=True, right_index=True)
# df_com.loc[(slice(None), 200001), :]

# for value in df_com.index.get_level_values(0):
#    print('%s: %s' % (value, str(len(df_com.loc[value]))))
    
df_com = df_com.reset_index().groupby(['ST', 'LEAID']).sum()  

x = np.array(df_com.index.get_level_values(0))
lst_states = np.unique(x)

# for value in lst_states:
#    print('%s: %s' % (value, str(len(df_com.loc[value]))))
    
df_com = df_com[df_com['TOTALREV'] > 0]

df_com['Per pupil'] = df_com['TOTALREV'] / df_com['Total']

df_com['Maj Min'] = df_com[list(col_mem)].sum(axis=1) > df_com['White']

df_com.sort_values(by=['Per pupil'])
```

```python
def plot_func(a, f):
    plt.figure(2)
    x = np.linspace(0, 2*np.pi, num=1000)
    y = a*np.sin(1/f*x)
    plt.plot(x,y)
    plt.ylim(-1.1, 1.1)
    plt.title('a sin(f)')
    plt.show()

interactive_plot = interactive(plot_func, a=(-1,0,0.1), f=(0.1, 1))
output = interactive_plot.children[-1]
output.layout.height = '300px'
interactive_plot
```
